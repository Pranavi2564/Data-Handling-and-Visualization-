```python
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')

import numpy as np
print("step 1")
```

```python
fig=plt.figure()
ax=plt.axes()
ax.grid()
```

```python
fig=plt.figure()
ax=plt.axes()

x=np.linspace(0,10,1000)

ax.plot(x,np.sin(x))
ax.plot(x,np.cos(x))
ax.set_title('Simple Plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()

x=np.linspace(0,10,1000)

ax.plot(x,np.sin(x), label='sin',color='red')
ax.plot(x,np.cos(x),label='cos',color='g')
ax.set_title('Simple Plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
```

```python
fig=plt.figure()
ax=plt.axes()

x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x),label = 'sin')
ax.plot(x,np.cos(x),label = 'cos')
ax.set_title('Simple Plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.legend()
ax.plot(x,np.tan(x),label = 'tan')
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()

x=np.linspace(0,10,1000)

ax.plot(x,np.sin(x),label='sin',linestyle='dashed')
ax.plot(x,np.cos(x),label='cos',linestyle='dotted')
ax.plot(x,np.sin(x+1),label='cos',linestyle='dashdot')
ax.set_title('Multiple Lines')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.legend();
```

```python
from matplotlib import pyplot as plt
import numpy as np
fig=plt.figure()
ax=plt.axes()
ax.grid()

x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x))

ax.set_xlim(-5,15)
ax.set_ylim(-3,3)
```

```python
fig,axs=plt.subplots(2,2,figsize=(10,6))
x=np.linspace(0,10,1000)
axs[0,0].plot(x,np.sin(x))
axs[0,1].plot(x,np.cos(x),color='maroon')
axs[1,0].plot(x,np.sin(x+2),color='blue')
axs[1,1].plot(x,np.sin(x+4),color='green')
```

```python
data={'Apple':10,'Orange':15,'Lemon':5,'Lime':20}
names=list(data.keys())
values=list(data.values())
fig=plt.figure()
ax=plt.axes()
ax.grid(linestyle='--',color='0.85')
ax.bar(names,values)

ax.bar(names,values,color='PINK',width=0.35)
ax.set_xlabel('Fruit')
ax.set_ylabel('Quantity')
ax.set_title('FRUIT QUANTITIES')
```

```python
data={'Apple':10,'Orange':15,'Banana':13,'Pear':9,'Pomegranate':11,'Lemon':5,'Lime':20}
names=list(data.keys())
values=list(data.values())
fig=plt.figure()
ax=plt.axes()
ax.grid(linestyle='--',color='0.85')
ax.barh(names,values)
ax.set_xlabel('Fruit')
ax.set_ylabel('Quantity')
ax.set_title('FRUIT QUANTITIES')
```
